```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time

def func4(a, b):
    count = 0
    for i in a:
        if i in b:
            count += 1
    return count

start = time.time()
print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
end = time.time()

print(end - start)
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def find_max(numbers):
    max_value=numbers[0]
    for n in numbers:
        if n > max_value:
            max_value = n
            return max_value
def find_min(numbers):
    min_value = numbers[0]
    for n in numbers:
        if n < min_value:
            min_value = n
    return min_value

print(find_min([3, 7, 2, 9, 4]))
print(find_max([3, 7, 2, 9, 4]))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def counting_even(numbers):
    even_count = 0
    for n in numbers:
        if n % 2 == 0:
            even_count += 1
    return even_count

def counting_odd(numbers):
    odd_count = 0
    for n in numbers:
        if n % 2 != 0:
            odd_count += 1
    return odd_count

numbers = [3, 7, 2, 9, 4]

print("Even count:", counting_even(numbers))
print("Odd count:", counting_odd(numbers))

```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1
arr = [10, 25, 30, 45, 60]
key = 45

result = linear_Search(arr, key)

if result != -1:
    print("Element found at index", result)
else:
    print("Element not found")
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1
arr = [1, 3, 5, 9, 10]
key = 10
result = binary_search(arr, key)
print(result)
```
